# Voice processings

```python
def m4a_processing(file_dir, file_name, format, destination_dir):
  track = AudioSegment.from_file(os.path.join(*[file_dir,f'{file_name}{format}']), format="m4a")
  track = track.set_frame_rate(44100)
  silence_duration = silence.detect_leading_silence(track)
  track = track[silence_duration:]
  track = track[:1000]
  track.export(f'{destination_dir}/{file_name}.wav', format="wav")
```

```python
def stereo_to_mono(file_dir, file_name, format, destination_dir):
  track = AudioSegment.from_file(os.path.join(*[file_dir,f'{file_name}{format}']), format="wav")
  if(track.channels == 2):
    track = track.set_channels(1)
  print(track.channels)
  track.export(f'{destination_dir}/{file_name}.wav', format="wav")
```

```python
def bulk_processing(root_dir, dest_dir):
  for sub_dir, dirs, files in os.walk(root_dir):
    for file in files:
      sub = sub_dir.split('/')[-1]
      file_name = os.path.splitext(file)[0]
      file_format = os.path.splitext(file)[1]
      dest_sub_dir = os.path.join(dest_dir, sub)
      print(file_name)
      print(file_format)
      print(dest_sub_dir)
      m4a_processing(sub_dir, file_name, file_format, dest_sub_dir)
```

```python
def bulk_stereo_to_mono_conversion(root_dir, dest_dir):
  for sub_dir, dirs, files in os.walk(root_dir):
    sub = sub_dir.split('/')[-1]
    dest_sub_dir = os.path.join(dest_dir, sub)
    os.makedirs(dest_sub_dir)
    for file in files:
      file_name = os.path.splitext(file)[0]
      file_format = os.path.splitext(file)[1]
      print(file_name, file_format, dest_sub_dir)
      stereo_to_mono(sub_dir, file_name, file_format, dest_sub_dir)
```
